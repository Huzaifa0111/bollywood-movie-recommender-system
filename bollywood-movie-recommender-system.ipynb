```python
import numpy as np
import pandas as pd
```

```python
bollywood_crew_1950_to_2019 = pd.read_csv('./1950-2019/bollywood_crew_1950-2019.csv')
bollywood_crew_data_1950_to_2019 = pd.read_csv('./1950-2019/bollywood_crew_data_1950-2019.csv')
bollywood_full_1950_to_2019 = pd.read_csv('./1950-2019/bollywood_full_1950-2019.csv')
```

```python
bollywood_crew_1950_to_2019.rename(columns={'directors':'crew_id'},inplace=True)
```

```python
bollywood_crew_1950_to_2019.head()
```

```python
bollywood_crew_1950_to_2019 = bollywood_crew_1950_to_2019.merge(bollywood_crew_data_1950_to_2019,on='crew_id')
```

```python
bollywood_crew_1950_to_2019.head()
```

```python
new_df = bollywood_full_1950_to_2019.merge(bollywood_crew_1950_to_2019,on='imdb_id')
```

```python
new_df.head()
```

```python
new_df.info()
```

```python
# only going to store this columns
# imdb_id
# original_title
# poster_path
# genres
# summary
# actors
# tagline           
# name_x
```

```python
new_df = new_df[['imdb_id','original_title','poster_path','tagline','genres','summary','actors','name']]
```

```python
new_df = new_df.drop_duplicates()
```

```python
new_df.isnull().sum()
```

```python
new_df.info()
```

```python
new_df.drop_duplicates(subset=["actors"], inplace=True)
```

```python
new_df.rename(columns={'name':'director'},inplace=True)
```

```python
new_df.head()
```

```python
new_df['genres'] = new_df['genres'].apply(lambda x:x.replace("|"," "))
```

```python
new_df['actors'] = new_df['actors'].apply(lambda x:str(x).replace('|',',').replace(' ','').replace(',',' '))
```

```python
new_df['actors'] = new_df['actors'].str.split().apply(lambda x: ' '.join(x[:2]))
```

```python
new_df['director'] = new_df['director'].apply(lambda x:x.replace(" ",""))
```

```python
# new_df['actors'] = new_df['actors'].apply(lambda x:x.split())
```

```python
# new_df['director'] = new_df['director'].apply(lambda x:x.split())
```

```python
# new_df['genres'] = new_df['genres'].apply(lambda x:x.split())
```

```python
# new_df['summary'] = new_df['summary'].apply(lambda x:x.split())
```

```python
new_df.head()
```

```python
new_df['poster_path'].fillna('',inplace=True)
```

```python
new_df['tags'] = new_df['actors'] + new_df['director'] + new_df['genres'] + new_df['summary']
```

```python
new_df = new_df[['imdb_id','original_title','genres','tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('Loving')
```

```python
new_df.iloc[0].tags
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y) 
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.iloc[0].tags
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(enumerate(list(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['original_title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].genres)
        print(new_df.iloc[i[0]].original_title)
    
```

```python
new_df[new_df['original_title'].str.contains('Uri: The Surgical Strike')]
```

```python
recommend('Uri: The Surgical Strike')
```
